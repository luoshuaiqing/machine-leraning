```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data_set_path = "../data/airfoil_self_noise.dat"

head_set = [
    "Frequency",
    "Angle_of_attack",
    "Chord_length",
    "Free_stream_velocity",
    "Suction_side_displacement_thickness",
    "Scaled_sound_pressure_level"
]
df = pd.read_csv(data_set_path, names=head_set, sep="\t")
print("# of rows:", df.shape[0])
print("# of columns:", df.shape[1])
df

# TODO: what do rows and columns represent?


# The six columns represent the following, respectively:
# 1. Frequency, in Hertzs.
# 2. Angle of attack, in degrees.
# 3. Chord length, in meters.
# 4. Free-stream velocity, in meters per second.
# 5. Suction side displacement thickness, in meters.
# 6. Scaled sound pressure level, in decibels. (is also the only output)

# The 1503 rows represent 1503 individual test or observatoins about different size NACA 0012 airfoils
```

b) iii. What are the mean, the median, range, first and third quartiles, and interquartile ranges of 
each of the variables in the dataset? Summarize them in a table.

```python
table_df = []
df.describe()
for column in df.columns:
    data = [column]
    data.append(df.describe()[column]["mean"])
    data.append(df[column].median())
    data.append(df.describe()[column]["max"] - df.describe()[column]["min"])
    data.append(df.describe()[column]["25%"])
    data.append(df.describe()[column]["75%"])
    data.append(df.describe()[column]["75%"] - df.describe()[column]["25%"])
    table_df.append(data)
    
table = pd.DataFrame(table_df, columns=["variable", "mean", "median", "range", "first quartile", "third quartile",
                                       "interquartile range"])
table
```

c) For each predictor, fit a simple linear regression model to predict the response.
Describe your results. In which of the models is there a statistically significant association between the predictor and the response? Create some plots to back up your assertions. Are there any outliers2 that you would like to remove from your data for each of these regression tasks?

```python
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression

regressor_uni_coefficient = []
for header in head_set[:-1]:
    X = df[header].values.reshape(-1, 1)
    Y = df["Scaled_sound_pressure_level"].values.reshape(-1, 1)
    linear_regressor = LinearRegression()  # create object for the class
    linear_regressor.fit(X, Y)  # perform linear regression
    Y_pred = linear_regressor.predict(X)  # make predictions
    plt.scatter(X, Y)
    plt.plot(X, Y_pred, color='red') 
    plt.xlabel(header)
    plt.ylabel('Scaled_sound_pressure_level')
    plt.show()
#     Adding coefficients of each regressor to an array
    regressor_uni_coefficient.append(linear_regressor.coef_)


outlier_list = []
for header in head_set[:-1]:
    X = df[header].values.reshape(-1, 1)
    X = sm.add_constant(X)
    Y = df["Scaled_sound_pressure_level"].values.reshape(-1, 1)
    model = sm.OLS(Y, X).fit()
    predictions = model.predict(X) 
 
    print_model = model.summary()
#     print_model = model.summary2()
    print(header)
    print(print_model)
#     print(print_model.tables[1]["Coef."][0])
    print("\n\n\n")
    # Outliers:
    outliers = model.outlier_test()[:,0]
    cnt = 0
    for index in range(len(outliers)):
        if abs(outliers[index]) > 3:
            cnt += 1
    outlier_list.append([header, cnt])
    
print("outliers: ")
print(outlier_list)
# There are 1, 1, 4, 1, 1 number of outliers for the five predictors respectively
# Becuase p-value of all models are so small, there is a statistically significant association between all predictors


# Frequency: y = 127.3037 - 0.0009x
# Angle_of_attack: y = 126.0701 -0.1820x
# Chord_length: y = 127.2142 - 17.4170x
# Free_stream_velocity: y = 122.0172 + 0.0554x
# Suction_side_displacement_thickness: y = 126.6632 -164.0275x
```

h) Can you improve your model using possible interaction terms or nonlinear associations between the predictors and response? Train the regression model on a randomly selected 70% subset of the data with all predictors. Also, run a regression model involving all possible interaction terms and quadratic nonlinearities, and remove insignificant variables using p-values (be careful about interaction terms). Test both models on the remaining points and report your train and test MSEs.

```python
from sklearn.metrics import mean_squared_error
# training_set: select 70% subset of the data with all predictors
from sklearn.model_selection import train_test_split
training_set, test_set = train_test_split(df, test_size=0.3)


#train the data with all predictors
X = training_set.iloc[:,:-1]
X = sm.add_constant(X)
Y = training_set["Scaled_sound_pressure_level"]
model = sm.OLS(Y, X).fit()
# calculating train mse
prediction = model.predict(X)
mse = mean_squared_error(training_set["Scaled_sound_pressure_level"], prediction)
print("train mse with all predictors", mse, "\n")

# calculating test mse
X_test = test_set.iloc[:,:-1]
X_test = sm.add_constant(X_test)
prediction = model.predict(X_test) 
mse = mean_squared_error(test_set["Scaled_sound_pressure_level"], prediction)
print("test mse with all predictors", mse)

print("=============================\n\n\n")

poly = PolynomialFeatures(degree=5, interaction_only=True)
Frequency_squared = (training_set['Frequency']**2).to_numpy().reshape(-1, 1)
Angle_of_attack_squared = (training_set['Angle_of_attack']**2).to_numpy().reshape(-1, 1)
Chord_length_squared = (training_set['Chord_length']**2).to_numpy().reshape(-1, 1)
Free_stream_velocity = (training_set['Free_stream_velocity']**2).to_numpy().reshape(-1, 1)
Suction_side_displacement_thickness_squared = (training_set['Suction_side_displacement_thickness']**2).to_numpy().reshape(-1, 1)


X_poly = poly.fit_transform(training_set.iloc[:,:-1])
X_poly = np.append(X_poly, Frequency_squared, axis = 1)
X_poly = np.append(X_poly, Angle_of_attack_squared, axis = 1)
X_poly = np.append(X_poly, Chord_length_squared, axis = 1)
X_poly = np.append(X_poly, Free_stream_velocity, axis = 1)
X_poly = np.append(X_poly, Suction_side_displacement_thickness_squared, axis = 1)

Y = training_set["Scaled_sound_pressure_level"].values.reshape(-1, 1)
model = sm.OLS(Y, X_poly).fit()
print_model = model.summary2()
print(print_model)
print("\n\n=============================")
significant_columns = []
for index in range(len(print_model.tables[1]["P>|t|"])):
    if print_model.tables[1]["P>|t|"][index] <= 0.05:
        significant_columns.append(index)

print("significant predictors are inside following indexes in X_poly:\n", significant_columns)
X_poly = X_poly[:,significant_columns]


# retraining the data after removing the insignificant variables
model = sm.OLS(Y, X_poly).fit()
print_model = model.summary2()
print(print_model)
print("\n\n=============================")




# calculating train mse using the model after removing insignificant variables
prediction = model.predict(X_poly)
mse = mean_squared_error(training_set["Scaled_sound_pressure_level"], prediction)
print("train mse with all possible interaction terms and quadratic nonlinearities using the model after removing insignificant variables", mse, "\n")

# calculating test mse using the model after removing significant variables
Frequency_squared = (test_set['Frequency']**2).to_numpy().reshape(-1, 1)
Angle_of_attack_squared = (test_set['Angle_of_attack']**2).to_numpy().reshape(-1, 1)
Chord_length_squared = (test_set['Chord_length']**2).to_numpy().reshape(-1, 1)
Free_stream_velocity = (test_set['Free_stream_velocity']**2).to_numpy().reshape(-1, 1)
Suction_side_displacement_thickness_squared = (test_set['Suction_side_displacement_thickness']**2).to_numpy().reshape(-1, 1)

X_test = poly.fit_transform(test_set.iloc[:,:-1])
X_test = np.append(X_test, Frequency_squared, axis = 1)
X_test = np.append(X_test, Angle_of_attack_squared, axis = 1)
X_test = np.append(X_test, Chord_length_squared, axis = 1)
X_test = np.append(X_test, Free_stream_velocity, axis = 1)
X_test = np.append(X_test, Suction_side_displacement_thickness_squared, axis = 1)
X_test = X_test[:,significant_columns]
prediction = model.predict(X_test) 
mse = mean_squared_error(test_set["Scaled_sound_pressure_level"], prediction)
print("test mse with all possible interaction terms and quadratic nonlinearities using the model after removing insignificant variables", mse)
```

i) Perform k-nearest neighbor regression for this dataset using both normalized and raw features. Find the value of k âˆˆ {1,2,...,100} that gives you the best fit. Plot the train and test errors in terms of 1/k.

```python
from sklearn import neighbors


def KNN(k_neighbor, training_set, test_set):
    x_train = training_set.drop('Scaled_sound_pressure_level', axis=1)
    y_train = training_set['Scaled_sound_pressure_level']
    model = neighbors.KNeighborsRegressor(n_neighbors = k_neighbor)
    
    x_test = test_set.drop('Scaled_sound_pressure_level', axis=1)
    y_test = test_set['Scaled_sound_pressure_level']
    model.fit(x_train, y_train)  #fit the model
    pred = model.predict(x_test) #make prediction on test set
    return pred





min_error = 100000
train_error = []
test_error = []
best_k = 100000
for k in range(1, 101, 1):
    # Find test error!
    prediction = KNN(k, training_set, test_set)
    mse = mean_squared_error(test_set['Scaled_sound_pressure_level'], prediction)
    if mse < min_error:
        min_error = mse
        best_k = k
    test_error.append([1.0 / k, mse])
    
    # Find training error!
    prediction = KNN(k, training_set, training_set)
    mse = mean_squared_error(training_set['Scaled_sound_pressure_level'], prediction)
    train_error.append([1.0 / k, mse])

print("Raw features:\n")
print("best k: ", best_k)
print("lowest mse:", min_error)
        
test_error_data = pd.DataFrame(test_error, columns=['1/k', 'mse'])
training_error_data = pd.DataFrame(train_error, columns=['1/k', 'mse'])

# plotting test error data 
sns.scatterplot(x="1/k", y="mse", data=test_error_data)
plt.title('test mse for raw data')
plt.xlabel('1/k')
plt.ylabel('mse')

# plotting training error data 
sns.scatterplot(x="1/k", y="mse", data=training_error_data)
plt.title('training mse for raw data')
plt.xlabel('1/k')
plt.ylabel('mse')
```

```python
# using normalized data
normalized_df = (df - df.mean())/df.std()
training_set = (training_set - training_set.mean()) / training_set.std()
test_set = (test_set - test_set.mean()) / test_set.std()

min_error = 100000
train_error = []
test_error = []
best_k = 100000

for k in range(1, 101, 1):
    # Find test error!
    prediction = KNN(k, training_set, test_set)
    mse = mean_squared_error(test_set['Scaled_sound_pressure_level'], prediction)
    if mse < min_error:
        min_error = mse
        best_k = k
    test_error.append([1.0 / k, mse])
    
    # Find training error!
    prediction = KNN(k, training_set, training_set)
    mse = mean_squared_error(training_set['Scaled_sound_pressure_level'], prediction)
    train_error.append([1.0 / k, mse])
        
print("Normalized data:\n")
print("best k:", best_k)
print("lowest mse:", min_error)

test_error_data = pd.DataFrame(test_error, columns=['1/k', 'mse'])
training_error_data = pd.DataFrame(train_error, columns=['1/k', 'mse'])

# plotting test error data 
sns.scatterplot(x="1/k", y="mse", data=test_error_data)
plt.title('test mse for normalized data')
plt.xlabel('1/k')
plt.ylabel('mse')


# plotting training error data 
sns.scatterplot(x="1/k", y="mse", data=training_error_data)
plt.title('training mse for normalized data')
plt.xlabel('1/k')
plt.ylabel('mse')
```
